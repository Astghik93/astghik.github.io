```python
import urllib2 as urll
from lxml import html
```

```python
url = "https://www.youtube.com/feed/trending"
```

```python
response = urll.urlopen(url)
page = response.read()
tree = html.document_fromstring(page)
```

```python
videos_url = tree.cssselect('[class="yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link "] ')
```

```python
videos_title=tree.cssselect('[class="yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link "]')
```

```python
videos_duration=tree.cssselect('[class="video-time"]')
```

```python
videos_username = tree.cssselect('[class="g-hovercard yt-uix-sessionlink       spf-link "]')
```

```python
videos_views = tree.cssselect('[class="yt-lockup-meta-info"] ' )
```

```python
videos_list=[]
for a,b,c,d,e in zip(videos_url, videos_title, videos_duration,videos_username,videos_views):
    url="https://www.youtube.com"+a.get('href')
    title=b.get("title")
    dur=c.text
    duration=dur.replace("- Duration: ","")
    user=d.text
    views=e.cssselect('li')
    views1=views[1].text
    views2=views1.replace(" views","")
    videos_list.append({"URL": url, "Title": title, "Duration": duration, "User": user, "Views": views2})
```

```python
from pprint import pprint
```

```python
pprint(videos_list)
```

```python
with open('trends_videos1.json', 'w') as output:
    json.dump(videos_list, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv
```

```python
with open('trend_videos1.csv', 'w') as output2:
    for i in videos_list:
        writer = csv.writer(output2)
        for a, b in i.items():
            writer.writerow([a, b])
        writer.writerow([])
```
