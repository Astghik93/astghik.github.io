```python
import re
import requests
```

```python
from BeautifulSoup import *
```

```python
url = "https://www.youtube.com/feed/trending"
```

```python
response = requests.get(url)
data = response.text
```

```python
soup = BeautifulSoup(data)
```

```python
videos_url = soup.findAll('a', attrs={'class': 'yt-uix-tile-link yt-ui-ellipsis yt-ui-ellipsis-2 yt-uix-sessionlink      spf-link '})
```

```python
print videos_url
```

```python
for i in videos_url:
    print(i.get("title"))
```

```python
from pprint import pprint
```

```python
videos_duration = soup.findAll('span', attrs={'class': 'accessible-description'})
```

```python
videos_username = soup.findAll('a', attrs={'class': 'g-hovercard yt-uix-sessionlink       spf-link '})
```

```python
videos_views = soup.findAll('ul', attrs={'class':'yt-lockup-meta-info'})
```

```python
videos_list=[]
for a,b,c,d in zip(videos_url, videos_duration,videos_username,videos_views):
    url="https://www.youtube.com"+a.get('href')
    title=a.get('title')
    dur=b.text
    duration=dur.replace("- Duration: ","")
    user=c.text
    views=d.findAll('li')
    views1=views[1].text
    views2=views1.replace(" views","")
    videos_list.append({"URL": url, "Title": title, "Duration": duration, "User": user, "Views": views2})
```

```python
pprint(videos_list)
```

```python
import json
```

```python
with open('trend_videos.json', 'w') as output:
    json.dump(videos_list, output, sort_keys = True, indent = 4)
```

```python
import unicodecsv as csv
```

```python
with open('trend_videos.csv', 'w') as output2:
    for i in videos_list:
        writer = csv.writer(output2)
        for a, b in i.items():
            writer.writerow([a, b])
        writer.writerow([])
```
